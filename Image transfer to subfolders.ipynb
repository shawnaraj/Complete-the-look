```python
import shutil 
import os
import pandas as pd
```

```python
df = pd.read_csv('./imagenamestest.csv', names=['Filename','Class'])
```

```python
df = df[df['Class'].notna()]
df.head()
```

```python
labels = df.sort_values('Class')
```

```python
class_names = list(labels.Class.unique())
class_names
```

```python
for i in class_names:
    os.makedirs(os.path.join('test_',i))
```

```python
for i in class_names:
    os.makedirs(os.path.join('test_',i))
for c in class_names:
    for i in list(labels[labels['Class']== c]['Filename']):

        #create path to the image 
        get_image = os.path.join('test_',i)
        print(get_image)

        #If image has not already exist in the new folder create one        
        if not os.path.exists('test_/'+c+i):
            # move the image 
            move_image_to_cat = shutil.copy(get_image,'test_/'+c)
```
